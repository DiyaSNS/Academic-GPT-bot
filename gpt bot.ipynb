```python
from langchain_community.document_loaders import PyMuPDFLoader

pdf_path = "ugrulebook.pdf"
loader = PyMuPDFLoader(pdf_path)
documents = loader.load()

print(documents[0].page_content[:1000])
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Split each document into chunks of 1000 characters with 200 character overlap
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
docs = text_splitter.split_documents(documents)

print(f"Total Chunks: {len(docs)}")
print(docs[0].page_content[:500])
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain.vectorstores import Chroma

embedding = GoogleGenerativeAIEmbeddings(model="models/embedding-001")


vectorstore = Chroma.from_documents(documents=docs, embedding=embedding, persist_directory="rag_chroma_db")

# (Optional) Persist for reuse later
vectorstore.persist()
```
